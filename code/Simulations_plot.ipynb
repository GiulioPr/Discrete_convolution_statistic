```python
import numpy as np
import os
import seaborn as sns
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 'xx-large'
mpl.rcParams['axes.labelsize'] = 'xx-large'
mpl.rcParams['xtick.labelsize'] = 'x-large'
mpl.rcParams['ytick.labelsize'] = 'x-large'
mpl.rcParams['legend.frameon'] = True
mpl.rcParams['legend.framealpha'] = 0.5
mpl.rcParams['legend.fontsize'] = 'x-large'
mpl.rcParams['lines.linewidth'] = 1.5
mpl.rcParams['figure.subplot.wspace'] = 0.2
mpl.rcParams['figure.subplot.hspace'] = 0.1

path_proj = os.path.dirname(os.getcwd())+ '/'
path_csv = path_proj + 'csv/'
path_plot = path_proj + 'figures/'

print(path_proj)

#PARAMETERS
#Level of significance
alpha = 0.05
#Ranks to be tested/degrees of freedom, for the convolution statistics
ranks = np.array([1, 2], dtype=int)
#Set plotting style
sns.set_style("ticks")
```

```python
def plot_figure1(df_plot, row_params, row_labs, col_params, col_labs, x_labs, ranks, acr, alpha=0.05):
    fig, ax = plt.subplots(
        nrows=len(row_params), ncols=len(col_params), sharex='all', sharey='all',
        figsize=(3*(len(col_params)), 3*(len(row_params)))
    )
    dense_ls = ['--', '-']
    loose_ls = [':', '-.']
    for nr,rpar in enumerate(row_params):
        for nc,cpar in enumerate(col_params):
            index = df_plot.apply(lambda r: all([r[lab]==cpar[k] for k,lab in enumerate(col_labs)]
                                                +[r[lab]==rpar[k] for k,lab in enumerate(row_labs)]),
                                  axis=1)
            df = df_plot[index]
            x = np.ravel(df[x_labs].values)
            ax[nr,nc].plot([0, max(x)], [alpha, alpha], ls='-', linewidth=0.5, c='black')
            ax[nr,nc].set_xlim(left=min(x), right=max(x)+0.02)
            ax[nr,nc].plot(x, df['P'].values,
                           ls=(0, (3, 5, 1, 5, 1, 5)), c='black', label=r'$P^{'+acr+'}_m$')
            for rk in ranks:
                ax[nr,nc].plot(x, df['C'+str(rk)].values,
                               ls=dense_ls[rk-1], c='black', label=r'$C^{'+acr+'}_{{%d}m}$' % rk)
                ax[nr,nc].plot(x, df['Z'+str(rk)].values,
                               ls=loose_ls[rk-1], c='black', label=r'$Z^{'+acr+'}_{{%d}m}$' % rk)
            if nr == len(row_params) - 1:
                ax[nr,nc].set_ylim(bottom=0., top=1.)
            if nc == 0:
                ax[nr,nc].set_ylabel('; '.join([r'$n_{{{0:}}}$'.format(k+1)+'='+str(int(rpar[k]))
                                        for k in range(len(rpar))]), fontsize='x-large')
                ax[nr,nc].set_yticks(np.arange(0., 1.1, 0.25))
            if nr == 0:
                ax[nr,nc].set_title('p='+str(cpar[0])+'; q='+str(cpar[1]), fontsize='x-large')
            if nr == len(row_params)-1:
                ax[nr,nc].set_xticks(np.arange(0., max(x)+0.02, 0.1))
            if nr == 0 and nc == len(col_params)-1:
                ax[nr,nc].legend(loc=2, bbox_to_anchor=(0.975, 1))
    fig.text(x=-0.07, y=0.5, s='Proportion of rejections at '+r'$\alpha$='+str(alpha),
             va='center', ha='left', fontsize='xx-large', rotation='vertical')
    fig.text(x=0.5, y=0.05, s=r'$\rho$',
             va='bottom', ha='center', fontsize='x-large')
    fig.subplots_adjust(wspace=.2, hspace=.2)
    sns.despine(fig=fig)
    return fig, ax

###FIGURE 1
df_plot = pd.read_csv(path_csv+'Fig1.csv', index_col=False)

lst_pq = [(0.3, 0.8), (0.1, 0.9)]
lst_nn = [np.array([10., 10., 10.]), np.array([10., 20., 20.]), np.array([100., 100., 100.])]
lst_rho = np.arange(0., 0.4, 0.01)

row_params = lst_nn
row_labs = ['n1', 'n2', 'n3']
col_params = lst_pq
col_labs = ['X1_p', 'X2_q']
x_labs = ['rho']

#GOF        
gof_cols = [lab for lab in df_plot.columns if not('ed' in lab)]
df_temp = df_plot[gof_cols].rename({'Pgof':'P', 'Cgof_1':'C1', 'Cgof_2':'C2',
                                    'Zgof_1':'Z1', 'Zgof_2':'Z2'}, axis=1)
fig, ax = plot_figure1(df_temp, row_params, row_labs, col_params, col_labs,
                        x_labs, ranks, acr='GF', alpha=0.05)
fig.text(x=0.5, y=0.94, s='Goodness-of-fit',
             va='top', ha='center', fontsize='xx-large')
fig.text(x=-0.07, y=0.94, s='a', fontweight='bold', va='top', ha='left', fontsize='xx-large')
fig.savefig(path_plot+'Fig1_gof.pdf', bbox_inches='tight')

#ED
ed_cols = [lab for lab in df_plot.columns if not('gof' in lab)]
df_temp = df_plot[ed_cols].rename({'Ped':'P', 'Ced_1':'C1', 'Ced_2':'C2',
                                   'Zed_1':'Z1', 'Zed_2':'Z2'}, axis=1)
fig, ax = plot_figure1(df_temp, row_params, row_labs, col_params, col_labs,
                        x_labs, ranks, acr='ED', alpha=0.05)
fig.text(x=0.5, y=0.94, s='Equality in distribution',
             va='top', ha='center', fontsize='xx-large')
fig.text(x=-0.07, y=0.94, s='b', fontweight='bold', va='top', ha='left', fontsize='xx-large')
fig.savefig(path_plot+'Fig1_ed.pdf', bbox_inches='tight')
```

```python
def plot_figure2(df_plot, row_params, row_labs, col_params, col_labs, x_labs, ranks, rho, acr, alpha=0.05):
    fig, ax = plt.subplots(
        nrows=len(row_params), ncols=len(col_params), sharex='all', sharey='all',
        figsize=(3*(len(col_params)), 3*(len(row_params)))
    )
    dense_ls = ['--', '-']
    loose_ls = [':', '-.']
    for nr,rpar in enumerate(row_params):
        for nc,cpar in enumerate(col_params):
            index = df_plot.apply(lambda r: all([r[lab]==cpar[k] for k,lab in enumerate(col_labs)]
                                                +[r[lab]==rpar[k] for k,lab in enumerate(row_labs)]
                                                +[r.rho==rho]),
                                  axis=1)
            df = df_plot[index]
            x = np.ravel(df[x_labs].values)
            ax[nr,nc].plot([0, max(x)], [alpha, alpha], ls='-', linewidth=0.5, c='black')
            ax[nr,nc].set_xlim(left=min(x), right=max(x)+0.02)
            ax[nr,nc].plot(x, df['P'].values,
                           ls=(0, (3, 5, 1, 5, 1, 5)), c='black', label=r'$P^{'+acr+'}_m$')
            for rk in ranks:
                ax[nr,nc].plot(x, df['C'+str(rk)].values,
                               ls=dense_ls[rk-1], c='black', label=r'$C^{'+acr+'}_{{%d}m}$' % rk)
                ax[nr,nc].plot(x, df['Z'+str(rk)].values,
                               ls=loose_ls[rk-1], c='black', label=r'$Z^{'+acr+'}_{{%d}m}$' % rk)
            ax[nr,nc].set_xscale('symlog')
            ax[nr,nc].set_xlim(left=min(x))
            if rho==0.:
                ax[nr,nc].set_ylim(bottom=0., top=0.15)
            else:
                ax[nr,nc].set_ylim(bottom=0., top=1.)
            if nr == len(row_params)-1:
                ax[nr,nc].set_xticks(x)
            if nr == 0:
                ax[nr,nc].set_title('p='+str(cpar[0])+'; q='+str(cpar[1]), fontsize='x-large')
            if nc == 0 and rho == 0:
                ax[nr,nc].set_yticks(np.arange(0., 0.16, 0.05))
            if nc == 0 and rho == 0.25:
                ax[nr,nc].set_yticks(np.arange(0., 1.1, 0.25))
            if nc == 0 and rho == 0:
                txt_lst = ['' if k==1 else '2' for k in rpar]
                ax[nr,nc].set_ylabel('; '.join([r'$n_{{{0:}}}$'.format(k+1)+'='+txt_lst[k]+r'$m$'
                                                for k in range(len(rpar))]),
                                     fontsize='x-large')
            if nr == 0 and nc == len(col_params)-1 and rho == 0.25:
                ax[nr,nc].legend(loc=2, bbox_to_anchor=(0.975, 1))
    if rho == 0:
        fig.text(x=-0.07, y=0.5, s='Proportion of rejections at '+r'$\alpha$='+str(alpha),
                 va='center', ha='left', fontsize='xx-large', rotation='vertical')
    fig.text(x=0.5, y=0.975, s=r'$\rho =$'+str(rho),
             va='top', ha='center', fontsize='xx-large')
    fig.text(x=0.5, y=0.05, s=r'$m$',
             va='bottom', ha='center', fontsize='x-large')
    fig.subplots_adjust(wspace=.3, hspace=.2)
    sns.despine(fig=fig)
    return fig, ax

###FIGURE 2
df_plot = pd.read_csv(path_csv+'Fig2.csv', index_col=False)

lst_pq = [(0.3, 0.8), (0.1, 0.9)]
lst_rho = np.array([0., 0.25])
lst_n = np.array([10., 25., 50., 75., 100., 250., 500., 750., 1000.])
base_nn = np.array([[1., 1., 1.], [1., 2., 2.]])

row_params = base_nn
row_labs = ['n1', 'n2', 'n3']
col_params = lst_pq
col_labs = ['X1_p', 'X2_q']
x_labs = ['m']

#GOF        
gof_cols = [lab for lab in df_plot.columns if not('ed' in lab)]
df_temp = df_plot[gof_cols].rename({'Pgof':'P', 'Cgof_1':'C1', 'Cgof_2':'C2',
                                    'Zgof_1':'Z1', 'Zgof_2':'Z2'}, axis=1)
for rho in lst_rho:
    fig, ax = plot_figure2(df_temp[df_temp.rho==rho], row_params, row_labs, col_params, col_labs,
                            x_labs, ranks, acr='GF', alpha=0.05, rho=rho)
    if rho == 0.:
            fig.text(x=-0.07, y=0.975, s='a', fontweight='bold', va='top', ha='left', fontsize='xx-large')
    elif rho == 0.25:
        fig.text(x=0.02, y=0.975, s='b', fontweight='bold', va='top', ha='left', fontsize='xx-large')
    fig.savefig( path_plot+'Fig2_gof'+str(rho).replace('.','')+'.pdf', bbox_inches='tight')

#ED
ed_cols = [lab for lab in df_plot.columns if not('gof' in lab)]
df_temp = df_plot[ed_cols].rename({'Ped':'P', 'Ced_1':'C1', 'Ced_2':'C2',
                                   'Zed_1':'Z1', 'Zed_2':'Z2'}, axis=1)
for rho in lst_rho:
    fig, ax = plot_figure2(df_temp[df_temp.rho==rho], row_params, row_labs, col_params, col_labs,
                               x_labs, ranks, acr='ED', alpha=0.05, rho=rho)
    if rho == 0.:
        fig.text(x=-0.07, y=0.975, s='c', fontweight='bold', va='top', ha='left', fontsize='xx-large')
    elif rho == 0.25:
        fig.text(x=0.02, y=0.975, s='d', fontweight='bold', va='top', ha='left', fontsize='xx-large')
    fig.savefig(path_plot+'Fig2_ed'+str(rho).replace('.','')+'.pdf', bbox_inches='tight')
```

```python
def plot_figure3(df_plot, row_params, row_labs, col_params, col_labs, x_labs, ranks, acr, alpha=0.05):
    fig, ax = plt.subplots(
        nrows=len(row_params), ncols=len(col_params), sharex='all', sharey='col',
        figsize=(3*(len(col_params)), 3*(len(row_params)))
    )
    dense_ls = ['--', '-']
    loose_ls = [':', '-.']
    for nr,rpar in enumerate(row_params):
        for nc,cpar in enumerate(col_params):
            index = df_plot.apply(lambda r: all([r[lab]==cpar[k] for k,lab in enumerate(col_labs)]
                                                +[r[lab]==rpar[k] for k,lab in enumerate(row_labs)]),
                                  axis=1)
            df = df_plot[index]
            x = np.ravel(df[x_labs].values)
            step = x[1] - x[0]
            xx_f = np.arange(0.01, 0.25 + step, step)
            xx_n = np.arange(0.25, 0.5, step)
            ax[nr,nc].plot([0, max(xx_n)], [alpha, alpha], ls='-', linewidth=0.5, c='black')
            ax[nr,nc].set_xlim(left=min(xx_f), right=max(xx_n)+0.02)
    
            for xx,p_loc in zip([xx_f, xx_n], ['far','near']):
                df_p_loc = df[df.where_p==p_loc]
                if p_loc == 'near':
                    ax[nr,nc].plot(xx, df_p_loc['P'].values,
                                   ls=(0, (3, 5, 1, 5, 1, 5)), c='black', label=r'$P^{'+acr+'}_m$')
                    for rk in ranks:
                        ax[nr,nc].plot(xx, df_p_loc['C'+str(rk)].values,
                                       ls=dense_ls[rk-1], c='black', label=r'$C^{'+acr+'}_{{%d}m}$' % rk)
                        ax[nr,nc].plot(xx, df_p_loc['Z'+str(rk)].values,
                                       ls=loose_ls[rk-1], c='black', label=r'$Z^{'+acr+'}_{{%d}m}$' % rk)
                else:
                    ax[nr,nc].plot(xx, df_p_loc['P'].values,
                                   ls=(0, (3, 5, 1, 5, 1, 5)), c='black')
                    for rk in ranks:
                        ax[nr,nc].plot(xx, df_p_loc['C'+str(rk)].values,
                                       ls=dense_ls[rk-1], c='black')
                        ax[nr,nc].plot(xx, df_p_loc['Z'+str(rk)].values,
                                       ls=loose_ls[rk-1], c='black')
            if nc == 0:
                ax[nr,nc].plot([0.25,0.25], [0.,0.25], ls='-', linewidth=0.5, c='black')
                ax[nr,nc].set_ylabel('; '.join([r'$n_{{{0:}}}$'.format(k+1)+'='+str(int(rpar[k]))
                                        for k in range(len(rpar))]), fontsize='x-large')
                ax[nr,nc].set_yticks(np.arange(0., 1.1, 0.1))
                ax[nr,nc].set_ylim(bottom=0., top=0.25)
            if nc == len(col_params)-1:
                ax[nr,nc].set_ylim(bottom=0., top=1.)
                ax[nr,nc].plot([0.25,0.25], [0.,1.], ls='-', linewidth=0.5, c='black')
                ax[nr,nc].set_yticks(np.arange(0., 1.1, 0.25))
            if nr == 0:
                ax[nr,nc].set_title(r'$\rho$ = '+str(cpar[0]), fontsize='x-large')
            if nr == len(row_params)-1:
                ax[nr,nc].set_xticks(np.arange(0., 0.51, 0.125))
                ax[nr,nc].set_xticklabels(['0.0','0.125','0.25'+'\n'+'0.55','0.675','0.8'], rotation='vertical')
            if nr == 0 and nc == len(col_params)-1:
                ax[nr,nc].legend(loc=2, bbox_to_anchor=(0.975, 1))
    fig.text(x=-0.07, y=0.5, s='Proportion of rejections at '+r'$\alpha$='+str(alpha),
             va='center', ha='left', fontsize='xx-large', rotation='vertical')
    fig.text(x=0.5, y=0.05, s=r'$p$',
             va='bottom', ha='center', fontsize='x-large')
    fig.subplots_adjust(wspace=.3, hspace=.2)
    sns.despine(fig=fig)
    return fig, ax

###FIGURE 3
df_plot = pd.read_csv(path_csv+'Fig3.csv', index_col=False)

q = 0.8
lst_nn = [np.array([10., 10., 10.]), np.array([10., 20., 20.]), np.array([100., 100., 100.])]
lst_rho = np.array([[0.], [0.25]])
#(p CLOSE TO q)
lst_p_near = np.arange(0.55, 0.8, 0.01)
#(p FAR FROM q)
lst_p_far = np.arange(0.01, 0.251, 0.01)

row_params = lst_nn
row_labs = ['n1', 'n2', 'n3']
col_params = lst_rho
col_labs = ['rho']
x_labs = ['X1_p']

#GOF        
gof_cols = [lab for lab in df_plot.columns if not('ed' in lab)]
df_temp = df_plot[gof_cols].rename({'Pgof':'P', 'Cgof_1':'C1', 'Cgof_2':'C2',
                                    'Zgof_1':'Z1', 'Zgof_2':'Z2'}, axis=1)
fig, ax = plot_figure3(df_temp, row_params, row_labs, col_params, col_labs,
                        x_labs, ranks, acr='GF', alpha=0.05)
fig.text(x=0.5, y=0.94, s='Goodness-of-fit',
             va='top', ha='center', fontsize='xx-large')
fig.text(x=-0.07, y=0.94, s='a', fontweight='bold', va='top', ha='left', fontsize='xx-large')
fig.savefig(path_plot+'Fig3_gof.pdf', bbox_inches='tight')

#ED
ed_cols = [lab for lab in df_plot.columns if not('gof' in lab)]
df_temp = df_plot[ed_cols].rename({'Ped':'P', 'Ced_1':'C1', 'Ced_2':'C2',
                                   'Zed_1':'Z1', 'Zed_2':'Z2'}, axis=1)
fig, ax = plot_figure3(df_temp, row_params, row_labs, col_params, col_labs,
                        x_labs, ranks, acr='ED', alpha=0.05)
fig.text(x=0.5, y=0.94, s='Equality in distribution',
             va='top', ha='center', fontsize='xx-large')
fig.text(x=-0.07, y=0.94, s='b', fontweight='bold', va='top', ha='left', fontsize='xx-large')
fig.savefig(path_plot+'Fig3_ed.pdf', bbox_inches='tight')
```
